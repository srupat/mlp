```python
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV
```

```python
X, y = load_wine(return_X_y=True, as_frame=True)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
```

```python
dt = DecisionTreeClassifier(random_state=1)
```

```python
param_grid = {
    'criterion': ['entropy', 'gini'],
    'splitter': ['random', 'best'],
    'min_samples_leaf': [2, 4, 6, 8, 10],
    'max_depth': [3, 4, 5, 6]
}

model = GridSearchCV(dt, param_grid, cv=4)
model.fit(X_train, y_train)
```

```python
score = model.score(X_test, y_test)
print(score)
```

```python
model.best_estimator_
```

```python
from sklearn.datasets import load_diabetes
```

```python
X, y = load_diabetes(return_X_y=True, as_frame=True)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
dr = DecisionTreeRegressor(random_state=1, criterion='squared_error', splitter='random', max_leaf_nodes=10)
```

```python
dr.fit(X_train, y_train)
```

```python
dr.score(X_test, y_test)
```

```python
dr.score(X_train, y_train)
```

```python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 10))
plot_tree(dr, filled=True)
plt.show()
```

```python
dr1 = DecisionTreeRegressor(max_depth=4, random_state=1)
```

```python
param_grid = {
    'criterion': ['squared_error', 'friedman_mse', 'absolute_error', 'poisson'],
    'splitter': ['random', 'best'],
    'max_features': [4,5,6,'sqrt', 'log2'],
    'ccp_alpha': [0.001,0.01,0.02,0.05]
}

model = GridSearchCV(dr1, param_grid, cv=4)
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
model.best_estimator_
```
