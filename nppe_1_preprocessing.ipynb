```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import MinMaxScaler, StandardScaler, OneHotEncoder
from sklearn.impute import SimpleImputer

?SimpleImputer
```

```python
df['RIVERSIDE'].unique()
```

```python
min_max_features = ['CRIM', 'ZN', 'POLINDEX', 'DIS', 'HIGHWAYCOUNT', 'TAX', 'PTRATIO', 'IMM', 'BPL']
pipe_median = Pipeline([("median_impute", SimpleImputer(strategy="median", missing_values=-1)),
                        ("scaler", MinMaxScaler())])

pipe_mean = Pipeline([("mean_impute", SimpleImputer(strategy="mean", missing_values=-2)),
                        ("scaler", MinMaxScaler())])

pipe_mode = Pipeline([("mode_impute", SimpleImputer(strategy="most_frequent", missing_values='UNKNOWN')),
                        ("encode", OneHotEncoder())])

ct = ColumnTransformer(
    [("min_max_scaling", MinMaxScaler(), min_max_features),
     ('standard_scaling', StandardScaler(), ['INDUS']),
     ('pipe_median', pipe_median, ['RM']),
     ('mean_pipe', pipe_mean, ['AGE']),
     ('mode_pipe', pipe_mode, ['RIVERSIDE'])]
)

transformed = ct.fit_transform(X_train)
```
